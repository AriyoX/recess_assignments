```python
# Removing wrong data from df1 
# Defining a custom function to validate and filter rows
def validate_row(row):
    if (row['Duration'] < 0) or (row['Pulse'] < 0) or (row['Maxpulse'] < 0) or (row['Calories'] < 0):
        return False
    if not (60 <= row['Pulse'] <= 160):
        return False
    if not (100 <= row['Maxpulse'] <= 220):
        return False
    # All checks pass, keep the row
    return True

df1_filtered = df1[df1.apply(validate_row, axis=1)]
df1_filtered = df1_filtered.dropna()
df1_filtered.head(30)
```

```python
# Removing wrong data in df2
df2_filtered = df2.dropna()
df2_filtered.head(10)
```

```python
# Removing wrong formats in df2
pattern = r'^\d+$'  # Matches any sequence of digits
df2_filtered = df2_filtered[df2_filtered['py-score'].str.match(pattern, na=False)]
df2_filtered = df2_filtered[df2_filtered['age'].str.match(pattern, na=False)]
df2_filtered.head(30)
```

```python
# Removing wrong formats in df1
df1_filtered = df1_filtered[df1_filtered['Duration'].astype(str).str.match(r'^\d+$', na=False)]
df1_filtered = df1_filtered[df1_filtered['Pulse'].astype(str).str.match(r'^\d+$', na=False)]
df1_filtered = df1_filtered[df1_filtered['Maxpulse'].astype(str).str.match(r'^\d+$', na=False)]
df1_filtered = df1_filtered[df1_filtered['Calories'].astype(str).str.match(r'^\d+(\.\d+)?$', na=False)]
df1_filtered.head(30)
```

```python
# Removing duplicates in df1 and df2
df1_filtered = df1_filtered.drop_duplicates()
df2_filtered = df2_filtered.drop_duplicates()
```
